```python
#Importación de librerías

import numpy as np
import matplotlib.pyplot as plt
```

#### **Función discreta**

```python
#Definimos los valores de la señal --- X[n]: Donde n define cada posición

n=np.array([-2,-1,0,1,2]) #Vector de posiciones
x=np.array([0,0,1,0,2]) #Valores de cada una de las posiciones (gorrito=posición cero)

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('X[n]')
plt.title('Señal discreta')
plt.grid()
```

#### **Función impulso unitario**

```python
def impulso_unitario(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n==n0).astype(int)
    return x,n

impulso_unitario(0,)
```

#### **Función escalón unitario**

```python
def escalon_unitario(n0,n1,n2):
    
    if n1 > n2:
        raise ValueError("n1 debe ser menor o igual que n2")
    
    #Vector de muestras
    n=np.arange(n1,n2+1)

    #Condición escalón
    u=np.where(n>=n0, 1,0)
    return n,u

n,u= escalon_unitario(0,-3,3)

print("n =", n)
print("u(n) =", u)

# Gráfica
plt.stem(n, u)
plt.title("Escalón Unitario u(n - n0)")
plt.xlabel("n")
plt.ylabel("u(n)")
plt.grid(True)
plt.show()
```

#### **Función rampa**

```python
def rampa(n0,n1,n2,m):

    n=np.arange(n1,n2+1)

    #Rampa
    r=np.where(n>=n0, m*(n-n0),0)
    return n,r
n,r=rampa(-5,-8,2,1)
plt.stem(n,r)
```

#### **Exponenciales complejas**

```python
#Exponencial  compleja definida en un intervalo 𝑛1≤𝑛0≤𝑛2.

def func_comp(sigma, w0, n1, n2):

    # Vector de tiempo discreto
    n = np.arange(n1, n2+1)

    # Exponencial compleja
    x = np.exp((sigma + 1j*w0) * n)

    return n, x

#Implementación

n, x = func_comp(-0.2, np.pi/4, -10, 10)

plt.stem(n, np.real(x))
plt.xlabel('n')
plt.ylabel('Re[x(n)]')
plt.title('Señal compleja')
plt.grid()
plt.show()
```

#### **Función senosoidal discreta**

y(n)=Asin(ω0​n+ϕ)

```python
def sin_signal(n1, n2, A, f, phi=0):

    # Vector 
    n = np.arange(n1, n2+1)

    # Frecuencia angular
    w0 = 2*np.pi*f

    # Señal sinusoidal
    y = A * np.sin(w0*n + phi)

    return n, y

#Implementación
n, y = sin_signal(-10, 10, 2, 0.1)

plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y(n)')
plt.title('Señal senosoidal')
plt.grid()
plt.show()
```

```python
#Secuencia 1: 𝑥1(𝑛)= 3𝛿(𝑛+1)+5𝛿(𝑛+3)+3𝛿(𝑛+2)+3𝛿(𝑛)+ 𝛿(𝑛)

# Vector de muestras
n = np.arange(-3,1)

# Inicializar señal en ceros
x = np.zeros(len(n))

# Asignar valores manualmente
x[n == -3] = 5
x[n == -2] = 3
x[n == -1] = 3
x[n == 0]  = 4

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x1(n)')
plt.title('Sinusoide 1')
plt.grid()
plt.show()
```
